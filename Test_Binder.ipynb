### First, begin my importing required modules

```python
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
```

### Use widgets to make interactive plot. 

```python
%matplotlib notebook

x = np.linspace(-2 * np.pi, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(phase_arg = 1.0, amplitude = 1):
    line.set_ydata(amplitude * np.sin(phase_arg * x))
    fig.canvas.draw_idle()
    ax.set_ylim([-5, 5])

interact(update);
```
